```python
import matplotlib.pyplot as plt
import torch
import numpy as np
train_loss_epochs = torch.load("../Models/TrainingLossVLQ_TrainedVLQ_nLayers_5_nNodes_1024.data")
test_loss_epochs_1 = torch.load("../Models/TestLoss1VLQ_TrainedVLQ_nLayers_5_nNodes_1024.data")
test_loss_epochs_2 = torch.load("../Models/TestLoss2VLQ_TrainedVLQ_nLayers_5_nNodes_1024.data")

epochs = np.arange(1, len(test_loss_epochs_2)+1)
plt.plot(epochs, train_loss_epochs, label="Training  Loss")
plt.plot(epochs, test_loss_epochs_1, label="Testing  Loss (train+test set)")
plt.plot(epochs, test_loss_epochs_2, label="Testing  Loss (test only set)")
plt.xlabel("epochs")
plt.ylabel("mean squared relative error")
plt.yscale("log")
plt.legend()
#this_model = modelname.split('/')[-1]
plt.savefig("../plots/LossEpochs.png")
plt.show()
```
